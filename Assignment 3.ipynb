```python
# 1. Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, **): ")

if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "*":
    print(num1 * num2)
elif op == "/":
    if num2 != 0:
        print(num1 / num2)
    else:
        print("Cannot divide by zero")
elif op == "**":
    print(num1 ** num2)
else:
    print("Invalid operator")
```

```python
# 2. Check if numeric value in list

my_list = ["a", "b", 5, "c", 10, "d"]
found = False
for item in my_list:
    if type(item) == int or type(item) == float:
        found = True
        break
if found:
    print("Numeric value found")
else:
    print("No numeric value found")
```

```python
# 3. Add a key to a dictionary

my_dict = {"name": "Ali", "age": 20}
my_dict["city"] = "Karachi"
print(my_dict)
```

```python
# 4. Sum all numeric items in a dictionary

my_dict = {"a": 10, "b": 20, "c": 30, "d": "hello", "e": 40}
total = 0
for key in my_dict:
    if type(my_dict[key]) == int or type(my_dict[key]) == float:
        total = total + my_dict[key]
print(total)
```

```python
# 5. Identify duplicate values from list

my_list = [1, 2, 3, 2, 4, 5, 3, 6, 7, 5]
duplicates = []
for i in range(len(my_list)):
    for j in range(i + 1, len(my_list)):
        if my_list[i] == my_list[j] and my_list[i] not in duplicates:
            duplicates.append(my_list[i])
print("Duplicate values:", duplicates)
```

```python
# 6. Check if a given key exists in a dictionary

my_dict = {"name": "Ali", "age": 20, "city": "Karachi"}
key = input("Enter key to check: ")

if key in my_dict:
    print("Key exists")
else:
    print("Key does not exist")
```
